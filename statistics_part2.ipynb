### What is hypothesis testing in statistics
Hypothesis testing is a statistical method that uses sample data to evaluate a hypothesis about a population parameter. It helps determine whether to reject the null hypothesis based on the evidence provided by the sample.

### What is the null hypothesis, and how does it differ from the alternative hypothesis
The null hypothesis (H₀) states that there is no effect or no difference, while the alternative hypothesis (H₁) suggests that there is an effect or a difference.

### What is the significance level in hypothesis testing, and why is it important
The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. Common values are 0.05, 0.01. It defines the threshold for statistical significance.

### What does a P-value represent in hypothesis testing
A P-value is the probability of obtaining results at least as extreme as the observed results, assuming the null hypothesis is true.

### How do you interpret the P-value in hypothesis testing
If P-value ≤ α, reject the null hypothesis. If P-value > α, fail to reject the null hypothesis.

### What are Type 1 and Type 2 errors in hypothesis testing
- Type 1 Error: Rejecting a true null hypothesis (false positive).
- Type 2 Error: Failing to reject a false null hypothesis (false negative).

### What is the difference between a one-tailed and a two-tailed test in hypothesis testing
- One-tailed test checks for a deviation in one direction.
- Two-tailed test checks for deviations in both directions.

### What is the Z-test, and when is it used in hypothesis testing
Z-test is used to determine if there is a significant difference between sample and population means when population variance is known and sample size is large.

### How do you calculate the Z-score, and what does it represent in hypothesis testing
Z = (X̄ - μ) / (σ / √n), where X̄ is the sample mean, μ is the population mean, σ is the population std. dev., n is the sample size. It represents how many std deviations a value is from the mean.

### What is the T-distribution, and when should it be used instead of the normal distribution
T-distribution is used when the sample size is small (n < 30) and population standard deviation is unknown.

### What is the difference between a Z-test and a T-test
- Z-test: Known population variance, large sample size.
- T-test: Unknown population variance, small sample size.

### What is the T-test, and how is it used in hypothesis testing
T-test evaluates whether the means of two groups are statistically different from each other.

### What is the relationship between Z-test and T-test in hypothesis testing
Both test means; Z-test is a special case of T-test when sample size is large and variance is known.

### What is a confidence interval, and how is it used to interpret statistical results
A confidence interval gives a range of values for a population parameter, estimated from a sample, and indicates the reliability of the estimate.

### What is the margin of error, and how does it affect the confidence interval
Margin of error indicates the range within which the true population parameter is expected to lie. Wider intervals mean more uncertainty.

### How is Bayes' Theorem used in statistics, and what is its significance
Bayes' Theorem updates the probability of a hypothesis based on new evidence. It's significant in decision making and Bayesian inference.

### What is the Chi-square distribution, and when is it used
Chi-square distribution is used for categorical data to assess goodness-of-fit and independence between variables.

### What is the Chi-square goodness of fit test, and how is it applied
It tests whether the observed distribution of data fits an expected distribution.

### What is the F-distribution, and when is it used in hypothesis testing
F-distribution is used in ANOVA and in comparing two variances.

### What is an ANOVA test, and what are its assumptions
ANOVA tests whether there are significant differences between the means of three or more groups. Assumptions: independence, normality, equal variances.

### What are the different types of ANOVA tests
- One-way ANOVA
- Two-way ANOVA
- Repeated measures ANOVA

### What is the F-test, and how does it relate to hypothesis testing?
F-test compares two variances and is used in ANOVA to test group differences.

 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(size=1000, mean=0, std=1):
    # Generate data from a normal distribution
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    skewness = skew(data)
    kurt = kurtosis(data, fisher=True)  # fisher=True → normal distribution kurtosis = 0

    # Display the results
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis (Fisher): {kurt:.4f}")
    return skewness, kurt

# Example usage
calculate_skewness_kurtosis()
```

 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Print the results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Population Mean: {population_mean}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpret the result
    alpha = 0.05  # 95% confidence level
    if p_value < alpha:
        print("👉 Reject the null hypothesis: Significant difference.")
    else:
        print("✅ Fail to reject the null hypothesis: No significant difference.")

# Example usage
np.random.seed(42)  # For reproducibility
sample = np.random.normal(loc=102, scale=15, size=50)  # Sample with mean ~102
z_test(sample_data=sample, population_mean=100, population_std=15)
```

 Implement a one-sample Z-test using Python to compare the sample mean with the population mean


```python
import numpy as np
from scipy.stats import norm

# ✅ Step 1: Define the Z-test function
def one_sample_z_test(sample,
                      population_mean,
                      population_std,
                      alpha: float = 0.05,
                      tail: str = "two"):
    sample = np.asarray(sample)
    n = sample.size
    sample_mean = sample.mean()

    # Z-score calculation
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # P-value calculation based on test type
    if tail == "two":
        p = 2 * (1 - norm.cdf(abs(z)))
    elif tail == "right":
        p = 1 - norm.cdf(z)
    elif tail == "left":
        p = norm.cdf(z)
    else:
        raise ValueError("tail must be 'two', 'left', or 'right'")

    decision = "Reject H0" if p < alpha else "Fail to reject H0"
    return z, p, decision

# ✅ Step 2: Generate sample data
np.random.seed(42)  # For consistent output
sample_data = np.random.normal(loc=102, scale=15, size=60)  # Sample with mean around 102

# ✅ Step 3: Define population parameters
population_mean = 100   # Known population mean
population_std = 15     # Known population standard deviation

# ✅ Step 4: Perform Z-test
z_score, p_value, decision = one_sample_z_test(
    sample=sample_data,
    population_mean=population_mean,
    population_std=population_std,
    alpha=0.05,       # 5% significance level
    tail="two"        # Two-tailed test
)

# ✅ Step 5: Print the results
print("------ Z-Test Result ------")
print(f"Sample Mean : {np.mean(sample_data):.2f}")
print(f"Z-score     : {z_score:.3f}")
print(f"P-value     : {p_value:.4f}")
print(f"Decision    : {decision}")
```

Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Function to perform Z-test
def two_tailed_z_test(sample, population_mean, population_std, alpha=0.05):
    sample = np.asarray(sample)
    n = len(sample)
    sample_mean = np.mean(sample)

    # Z-score calculation
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Decision
    critical_value = norm.ppf(1 - alpha / 2)
    decision = "Reject H0" if abs(z) > critical_value else "Fail to reject H0"

    return z, p_value, critical_value, decision

# Step 2: Generate sample data
np.random.seed(42)
sample_data = np.random.normal(loc=102, scale=15, size=60)  # Sample ~ N(102, 15)

# Step 3: Population parameters
mu_0 = 100
sigma_pop = 15
alpha = 0.05

# Step 4: Perform Z-test
z_score, p_value, z_critical, decision = two_tailed_z_test(
    sample=sample_data,
    population_mean=mu_0,
    population_std=sigma_pop,
    alpha=alpha
)

# Step 5: Print Results
print("------ Z-Test Result ------")
print(f"Sample Mean : {np.mean(sample_data):.2f}")
print(f"Z-score     : {z_score:.3f}")
print(f"P-value     : {p_value:.4f}")
print(f"Critical Z  : ±{z_critical:.3f}")
print(f"Decision    : {decision}")

# Step 6: Visualization
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')

# Critical regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.3)

# Shade area under curve for z-score
plt.axvline(z_score, color='black', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.axvline(-z_critical, color='red', linestyle=':', label=f'Critical Z = ±{z_critical:.2f}')
plt.axvline(z_critical, color='red', linestyle=':')

plt.title("Two-Tailed Z-Test with Rejection Regions")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu_null=100, mu_alt=105, sigma=10, n=30, alpha=0.05):
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical z-value for two-tailed test
    z_critical = norm.ppf(1 - alpha / 2)

    # Critical sample means under null
    crit_low = mu_null - z_critical * se
    crit_high = mu_null + z_critical * se

    # Create range of x-values for plotting
    x = np.linspace(mu_null - 4*se, mu_alt + 4*se, 1000)

    # Null and alternative distributions
    y_null = norm.pdf(x, mu_null, se)
    y_alt = norm.pdf(x, mu_alt, se)

    plt.figure(figsize=(12, 6))

    # Plot both distributions
    plt.plot(x, y_null, label='Null Hypothesis (H₀)', color='blue')
    plt.plot(x, y_alt, label='Alternative Hypothesis (H₁)', color='green')

    # Shade Type I error regions (alpha)
    plt.fill_between(x, y_null, where=(x <= crit_low), color='red', alpha=0.3, label='Type I Error (α)')
    plt.fill_between(x, y_null, where=(x >= crit_high), color='red', alpha=0.3)

    # Shade Type II error region (beta)
    plt.fill_between(x, y_alt, where=(x > crit_low) & (x < crit_high), color='orange', alpha=0.3, label='Type II Error (β)')

    # Annotations
    plt.axvline(crit_low, color='black', linestyle='--', label=f'Critical Value = {crit_low:.2f}')
    plt.axvline(crit_high, color='black', linestyle='--')

    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend(loc='upper right')
    plt.grid(True)
    plt.show()

# Example usage
visualize_type1_type2_errors(mu_null=100, mu_alt=105, sigma=10, n=30, alpha=0.05)
```

 Write a Python program to perform an independent T-test and interpret the results

```python
import numpy as np
from scipy.stats import ttest_ind

# Step 1: Generate two independent samples
np.random.seed(42)

# Group A: Mean ≈ 100, std = 10
group_a = np.random.normal(loc=100, scale=10, size=50)

# Group B: Mean ≈ 105, std = 10
group_b = np.random.normal(loc=105, scale=10, size=50)

# Step 2: Perform Independent T-Test
t_stat, p_value = ttest_ind(group_a, group_b, equal_var=True)  # Assume equal variances

# Step 3: Print Results
print("------ Independent T-Test Result ------")
print(f"Group A Mean: {np.mean(group_a):.2f}")
print(f"Group B Mean: {np.mean(group_b):.2f}")
print(f"T-Statistic  : {t_stat:.3f}")
print(f"P-Value      : {p_value:.4f}")

# Step 4: Interpret Result
alpha = 0.05
if p_value < alpha:
    print("👉 Reject the null hypothesis: There is a significant difference between the group means.")
else:
    print("✅ Fail to reject the null hypothesis: No significant difference between the group means.")
```

 Perform a paired sample T-test using Python and visualize the comparison results


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

# Step 1: Simulate paired data (e.g., before vs after treatment)
np.random.seed(42)

# Before treatment (mean ≈ 100)
before = np.random.normal(loc=100, scale=10, size=30)

# After treatment (mean slightly improved)
after = before + np.random.normal(loc=2, scale=5, size=30)

# Step 2: Perform Paired Sample T-Test
t_stat, p_value = ttest_rel(before, after)

# Step 3: Print the results
print("------ Paired Sample T-Test Result ------")
print(f"Mean (Before): {np.mean(before):.2f}")
print(f"Mean (After) : {np.mean(after):.2f}")
print(f"T-Statistic   : {t_stat:.3f}")
print(f"P-Value       : {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("👉 Reject the null hypothesis: Significant change detected.")
else:
    print("✅ Fail to reject the null hypothesis: No significant change.")

# Step 4: Visualization
plt.figure(figsize=(10, 6))

# Line plot to show before vs after for each individual
for i in range(len(before)):
    plt.plot([0, 1], [before[i], after[i]], marker='o', color='gray', alpha=0.6)

# Mean lines
plt.plot([0]*len(before), before, 'bo', label='Before')
plt.plot([1]*len(after), after, 'go', label='After')

plt.xticks([0, 1], ['Before', 'After'])
plt.ylabel('Score / Value')
plt.title('Paired Sample Comparison (Before vs After)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

 Simulate data and perform both Z-test and T-test, then compare the results using Python


```python
import numpy as np
from scipy.stats import norm, t

# Step 1: Simulate sample data
np.random.seed(42)
sample = np.random.normal(loc=102, scale=15, size=25)  # Small sample (n < 30)

# Known population parameters for Z-test
population_mean = 100
population_std = 15
n = len(sample)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # sample std for T-test

# Step 2: Z-test (uses known population std)
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))
z_p = 2 * (1 - norm.cdf(abs(z_stat)))

# Step 3: T-test (uses sample std)
t_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
t_p = 2 * (1 - t.cdf(abs(t_stat), df=n-1))

# Step 4: Print Results
print("------ Z-Test vs T-Test ------")
print(f"Sample Mean        : {sample_mean:.2f}")
print(f"Sample Std (s)     : {sample_std:.2f}")
print(f"Population Std (σ) : {population_std}")
print(f"Sample Size (n)    : {n}")

print("\n✅ Z-Test Result (σ known):")
print(f"Z-Statistic         : {z_stat:.3f}")
print(f"P-value (Z-test)    : {z_p:.4f}")

print("\n✅ T-Test Result (σ unknown):")
print(f"T-Statistic         : {t_stat:.3f}")
print(f"P-value (T-test)    : {t_p:.4f}")

# Step 5: Interpretation
alpha = 0.05
print("\n🎯 Conclusion:")
if z_p < alpha:
    print("Z-Test     → Reject H0 (significant difference)")
else:
    print("Z-Test     → Fail to reject H0")

if t_p < alpha:
    print("T-Test     → Reject H0 (significant difference)")
else:
    print("T-Test     → Fail to reject H0")
```

 Write a Python function to calculate the confidence interval for a sample mean and explain its significance


```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
    - data: array-like sample data
    - confidence: desired confidence level (default = 0.95 for 95%)

    Returns:
    - (lower bound, upper bound) of the confidence interval
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # standard error
    df = n - 1
    t_crit = t.ppf((1 + confidence) / 2, df)  # critical t-value

    margin = t_crit * std_err
    return (mean - margin, mean + margin)
```

 Write a Python program to calculate the margin of error for a given confidence level using sample data


```python
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculate margin of error for a given sample and confidence level.

    Parameters:
    - data: array-like, sample observations
    - confidence: float, default 0.95 (for 95% confidence)

    Returns:
    - margin_of_error: float
    """
    data = np.array(data)
    n = len(data)
    sample_std = np.std(data, ddof=1)
    standard_error = sample_std / np.sqrt(n)
    df = n - 1
    t_critical = t.ppf((1 + confidence) / 2, df)
    margin_of_error = t_critical * standard_error
    return margin_of_error
```

 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process


```python
def bayes_theorem(prior, sensitivity, specificity):
    """
    Calculate posterior probability using Bayes' Theorem.

    Parameters:
    - prior: P(Disease) — prior probability (e.g. 0.01)
    - sensitivity: P(Pos | Disease) — true positive rate
    - specificity: P(Neg | No Disease) — true negative rate

    Returns:
    - posterior: P(Disease | Positive Test)
    """
    p_disease = prior
    p_no_disease = 1 - prior

    # Likelihoods
    p_pos_given_disease = sensitivity
    p_pos_given_no_disease = 1 - specificity  # False positive rate

    # Total probability of a positive test (evidence)
    p_pos = (p_pos_given_disease * p_disease) + (p_pos_given_no_disease * p_no_disease)

    # Bayes' Theorem
    posterior = (p_pos_given_disease * p_disease) / p_pos
    return posterior
```

 Perform a Chi-square test for independence between two categorical variables in Python


```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a contingency table
# Example: Gender vs Product Preference
data = pd.DataFrame({
    'Product_A': [30, 20],  # e.g., Male, Female
    'Product_B': [10, 25],
    'Product_C': [20, 15]
}, index=['Male', 'Female'])

print("Contingency Table:")
print(data)

# Step 2: Perform Chi-Square Test
chi2, p, dof, expected = chi2_contingency(data)

# Step 3: Print Results
print("\n--- Chi-Square Test Result ---")
print(f"Chi2 Statistic : {chi2:.3f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value        : {p:.4f}")

# Step 4: Interpretation
alpha = 0.05
if p < alpha:
    print("👉 Reject the null hypothesis: Variables are dependent (associated).")
else:
    print("✅ Fail to reject the null hypothesis: No significant association found.")

# Optional: Show expected frequencies
print("\nExpected Frequencies (if independent):")
expected_df = pd.DataFrame(expected, index=data.index, columns=data.columns)
print(expected_df.round(2))
```

 Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.


```python
import numpy as np
from scipy.stats import chisquare

# Step 1: Define observed and expected frequencies
# Example: Observed frequencies of dice outcomes
observed = np.array([18, 22, 20, 17, 19, 24])  # Sample counts for 6 faces of a die

# Step 2: Define expected frequencies (uniform distribution in this case)
total = np.sum(observed)
expected = np.array([total / 6] * 6)  # Uniform expectation

# Step 3: Perform Chi-Square Goodness-of-Fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 4: Print the results
print("🎲 Chi-Square Goodness-of-Fit Test")
print("----------------------------------")
print(f"Observed   : {observed}")
print(f"Expected   : {expected}")
print(f"Chi2 Stat  : {chi2_stat:.3f}")
print(f"P-value    : {p_value:.4f}")

# Step 5: Interpret the result
alpha = 0.05
if p_value < alpha:
    print("👉 Reject H₀: Observed data does not fit the expected distribution.")
else:
    print("✅ Fail to reject H₀: No significant difference between observed and expected.")
```
