Ques1. What is NumPy, and why is it widely used in Python?

Ans. NumPy (Numerical Python) is a library for numerical computations. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on them.

# Why it's widely used:

* Efficient memory storage
* Fast operations (vectorized computations)
* Supports linear algebra, Fourier transforms, random numbers, etc.

```python
import numpy as np
a = np.array([1, 2, 3])
print(a * 2)
```

Ques4. Explain the use of the groupby() method in Pandas.

Ques10. What is the significance of hierarchical indexing in Pandas?

Ques17. Explain the difference between apply() and map() in PandasA

Answer:

map() is for Series only, element-wise

apply() works on Series and DataFrames, applies a function

```python
df = pd.DataFrame({'A': [1, 2, 3]})
print(df['A'].map(lambda x: x * 2))
print(df.apply(np.sum))  # Sum per column
```

Ques19. How does Pandas simplify time series analysis?

Ans.
* Date parsing
* Resampling
* Rolling window
* Time zone handling

```python
dates = pd.date_range('20230101', periods=3)
df = pd.DataFrame({'val': [1, 2, 3]}, index=dates)
print(df.resample('D').mean())
```

Ques4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = sns.load_dataset('tips')

corr_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title("Correlation Heatmap of 'tips' Dataset")
plt.show()
```
