Ques6. Explain the concept of generators in Python and how they are defined.

Ans : Generators are special functions that allow you to iterate through a set of values one at a time.
1. They use yield instead of return.
2. State is saved between yields (does not start from beginning)




```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

for i in countdown(3):
    print(i)
```

Ques7. What are the advantages of using generators over regular functions?

Ans: Advantages of using generators over regular functions:

1. Memory Efficient: They do not store all values in memory at once.
2. Lazy Evaluation: Values are produced only when needed.
3. Faster in case of huge datasets: Avoids heavy memory consumption.

```python
def numbers():
    for i in range(1000000):
        yield i  # No memory overloading

g = numbers()
print(next(g))
print(next(g))
```

Ques10. What is the difference between map(), reduce(), and filter() functions in Python?

Ans: Difference between map(), reduce(), Filter() in Python:

1. map()	Applies a function to all elements.
2. filter()	Select elements meeting condition.
3. reduce()	Cumulatively applies a function to pairs

```python
from functools import reduce

nums = [1, 2, 3, 4]

print(list(map(lambda x: x * 2, nums)))
print(list(filter(lambda x: x % 2 == 0, nums)))
print(reduce(lambda x, y: x + y, nums))
```

Ques1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even(nums):
    even_sum = 0
    for num in nums:
        if num % 2 == 0:
            even_sum += num
    return even_sum

print("Sum of even number is:",sum_of_even([1, 2, 3, 4, 5, 6]))
```

Ques4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(num, end=" ")
```

Ques5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.


```python
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return self.a

fib = Fibonacci(7)
for num in fib:
    print(num, end=" ")
```

Ques6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(exponent):
    for i in range(exponent+1):
        yield 2 ** i

for num in powers_of_two(5):
    print(num)
```

Ques10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda ch: ch not in vowels, s))

print(remove_vowels("Beautiful World"))
```
